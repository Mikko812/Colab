```
#БЭТТЕРЫ. НОВАЯ ВЕРСИЯ
def Hits(hits,players):
  res=re.sub(r' [(]\d+[)]','', hits)
  res=res.replace('John Ryan Murphy','John Murphy').replace('Michael A. Taylor','Michael Taylor').replace('A.J. Pollock','Andy Pollock')
  res=res.replace('Michael A. Taylor','Michael Taylor')
  res=res.split(', ')
  hits=[]
  for el in res:
    if el[-2]==' ':
      hits.append(int(el[-1]))
    else:
      hits.append(1)
    
  res=list(map(lambda x: x.split(' ')[0][0]+'. '+x.split(' ')[1], res))
  result=[]
  for el in players:
    if el in res:
      i=res.index(el)
      result.append(hits[i])
    else:
      result.append(0)
  return result

def RunCreationPlayers(r):
  if r['AB']+r['BB']!=0:
    rc=round((r['H']+r['BB'])*r['TB']/(r['AB']+r['BB']),2)
    r['tRC']=rc
  else:
    r['tRC']=0
  return r

def NoHitter(bat3):
  cg=list(map(lambda x: 'TB - ' in x, bat3))
  if True in cg:
    return True   #no no-hitter
  else:
    return False

def batters(df):
  x=df['POS'].to_list().index('Totals')
  df.drop('POS',axis=1,inplace=True)
  bat1=df[:x]  #batters
  bat1.rename(columns={'Unnamed: 0':'Batter'},inplace=True)
  bat2=df[x:x+1] #Total
  bat3=df[x+2:]['AB'].to_list()  #stats
  
  bat2.loc[x,'AVG']=round(int(bat2['H'])/int(bat2['AB']),3)
  if NoHitter(bat3):   #если не nohitter
    double=list(filter(lambda x: x[:5]=='2B - ',bat3))
    triple=list(filter(lambda x: x[:5]=='3B - ',bat3))
    homerun=list(filter(lambda x: x[:5]=='HR - ',bat3))
    tb=list(map(lambda x: x if x[:5]=='TB - ' else '',bat3))+[',']
    tb=''.join(tb).split(' ')
    tb=list(map(lambda x: int(x[:-1]) if ',' in x else 0,tb))
    tb=sum(tb)
    bat3=list(filter(lambda x: 'TB - ' in x,bat3))[0]
  else:
    tb=0
  bat2.insert(9,'TB',tb)

  bat1['AVG']=list(map(lambda x: float(x),bat1['AVG'])) #AVG -> float
  for col in bat1.loc[:,'AB':'LOB']:                    #колонки -> int
    bat1[col]=list(map(int,bat1[col]))
  bat1.insert(8,'tAVG',0)
  bat1['tAVG']=round(bat1['H']/bat1['AB'],3)
  bat1.insert(10,'tRC',0)
  bat1.insert(8,'TB',0)
  bat1.insert(0,'Date',d_link[2:])
  bat1=bat1.drop_duplicates()
  totbase_dict=dict.fromkeys(bat1['Batter'],0)          #TB для каждого игрока
  if 'TB - ' in bat3:
    tb_all=bat3[5:].split(', ')
    for el in tb_all:
      el=el.replace('A.J. Pollock','A. Pollock').replace('T. La Stella','Tommy LaStella').replace('John Ryan Murphy','John Murphy').replace('Michael A. Taylor','Michael Taylor')
      el=el.replace('Michael A. Taylor','Michael Taylor').replace('Rafael Alejandro Marchan','Rafael Marchan').replace('Luis Alexander Basabe','Luis Basabe')
      sp1=el.find(' ')
      sp2=el.rfind(' ')
      name=el[0]+'. '+el[sp1+1:sp2]
      totbase=int(el[sp2+1:])
      totbase_dict[name]=totbase
  #print(totbase_dict)
  bat1['TB']=list(totbase_dict.values())
  bat1.fillna(0,inplace=True)
  bat1=bat1.apply(RunCreationPlayers,axis=1)
  if double!=[]:
    bat1['2B']=Hits(double[0][5:],bat1['Batter'].to_list())
  else:
    bat1['2B']=[0]*len(bat1)
  if triple!=[]:
    bat1['3B']=Hits(triple[0][5:],bat1['Batter'].to_list())
  else:
    bat1['3B']=[0]*len(bat1)
  if homerun!=[]:
    bat1['HR']=Hits(homerun[0][5:],bat1['Batter'].to_list())
  else:
    bat1['HR']=[0]*len(bat1)
  return bat1,bat2,tb

tmp_batters=pd.DataFrame()
all_batters=pd.DataFrame()
for tab in all_tabs:
  print(tab[0].loc[0,'Teams'],' - ',tab[0].loc[1,'Teams'])
  bat_A=tab[0].loc[0,'Teams']
  if bat_A in double_headers.keys():
    game=double_headers[bat_A]
    double_headers[bat_A]+=1
  else:
    game=''
  bat_A=teams_list[bat_A]
  bat_H=tab[0].loc[1,'Teams']
  bat_H=teams_list[bat_H]
  batters_A,total_A,tb_A=batters(tab[2])
  batters_H,total_H,tb_H=batters(tab[3])
  #display(batters_A,batters_H)
  tmp_batters=tmp_batters.append(total_A,sort=False)
  tmp_batters=tmp_batters.append(total_H,sort=False)
  batters_A.insert(2,'Game',game)
  batters_H.insert(2,'Game',game)
  batters_A=batters_A[['Date','Batter','Game','AB','R','H','2B','3B','HR','RBI','BB','SO','TB','tAVG','AVG','tRC']]
  batters_H=batters_H[['Date','Batter','Game','AB','R','H','2B','3B','HR','RBI','BB','SO','TB','tAVG','AVG','tRC']]
  if f"b-{bat_A}.csv" in list_mlb_dir:
    batters_A.to_csv(f"{path}b-{bat_A}.csv",index=False,header=False,mode='a')
  else:
    batters_A.to_csv(f"{path}b-{bat_A}.csv",index=False)
  if f"b-{bat_H}.csv" in list_mlb_dir:
    batters_H.to_csv(f"{path}b-{bat_H}.csv",index=False,header=False,mode='a')
  else:
    batters_H.to_csv(f"{path}b-{bat_H}.csv",index=False)

tmp_batters.drop(labels=['Unnamed: 0'],inplace=True,axis=1)
tmp_batters.columns=['AB','bR','bH','RBI','bBB','bSO','LOB','AVG','TB']
tmp_batters.reset_index(inplace=True,drop=True)
tmp_batters.reset_index(inplace=True)
display(tmp_batters)
```
